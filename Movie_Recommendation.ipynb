**Import Library**

```python
import pandas as pd
import numpy as np
```

**Import Dataset**

```python
df=pd.read_csv('https://github.com/Prarthana37/BIG-DATA-YBI/raw/461956d0da8013664eb09a6cf855f858416be0eb/netflix_data.csv')
df.head()
```

**Get Feature Selection**


```python
df_features = df[['title','cast','rating','duration']].fillna('')
```

```python
df_features.shape
```

```python
df_features
```

```python
x = df_features['title'] + ' ' + df_features['cast'] + ' ' + df_features['rating'] + ' ' + df_features['duration']
```

```python
x
```

```python
x.shape
```

**Get Feature Conversation to Tokens**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x = tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

**Get Similarity score using Cosine Similarity**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_Score = cosine_similarity(x)
```

```python
similarity_Score
```

```python
similarity_Score.shape
```

**Get Movie Name as Input from User and Vlidate for Closest  Spelling**

```python
favourite_movie_name= input("enter your favourite movie name:")
```

```python
all_title_list = df['title'].tolist()
```

```python
import difflib
```

```python
movie_recommendation = difflib.get_close_matches(favourite_movie_name,all_title_list)
print(movie_recommendation)
```

```python
close_match= movie_recommendation[0]
print(close_match)
```

```python
index_of_close_match_movie= df[df.title == close_match]['rating'].index[0]
print(index_of_close_match_movie)
```

```python
reccomendation_score = list(enumerate(similarity_Score[index_of_close_match_movie]))
print(reccomendation_score)
```

```python
len(reccomendation_score)
```

**Get All Movies based on their similarity Score**

```python
sorted_similar_movies = sorted(reccomendation_score,key= lambda x:x[1],reverse=True)
```

```python
print(sorted_similar_movies)
```

```python
print("top 30 movies suggested for you:\n")
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=df[df.index==index]['title'].values[0]
  if(i<31):
    print(i,'.',title_from_index)
    i+=1
```
